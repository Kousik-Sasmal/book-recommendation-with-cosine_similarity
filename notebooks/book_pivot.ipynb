```python
books.rename(columns={'Book-Title':'title','Book-Author':'author','Year-Of-Publication':'year','Publisher':'publisher'},inplace = True)
```

```python
books = books[['ISBN','title','author','year','publisher','Image-URL-M']]
books
```

```python
books['title'] = books['title'].str.replace("\\","")
books['title'] = books['title'].str.replace('"','')
```

```python
users.rename(columns={'User-ID':'user_id','Location':'location','Age':'age'},inplace = True)
users
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
ratings
```

```python
ratings.groupby('ISBN')['user_id'].count().sort_values(ascending=False)
```

```python
popular_df = ratings.groupby('ISBN').agg({'user_id':'count','rating':'mean'}).sort_values(ascending=False,by='user_id')
popular_df.rename(columns={'user_id':'num_ratings','rating':'avg_rating'},inplace=True)
```

```python
popular_df
```

```python
### fetch url for popular_df from `books.csv`

popular_df = popular_df.merge(books,on ='ISBN')
```

```python
popular_books = popular_df[0:4000][popular_df[0:4000]['avg_rating']>4]
popular_books
```

```python
popular_books = popular_books.reset_index(drop=True)
```

### set the min `number of ratings` by an user is 200

```python
ratings['user_id'].value_counts().shape
```

```python
ratings['user_id'].value_counts()[ratings['user_id'].value_counts()>200]
```

```python
x = ratings['user_id'].value_counts()
y = x[x>200]
y
```

```python
y.index
```

```python
ratings = ratings[ratings['user_id'].isin(y.index)]
ratings
```

```python
ratings_with_books= ratings.merge(books,on='ISBN')
ratings_with_books
```

```python
ratings_with_books['title'].nunique()
```

```python
ratings_with_books['ISBN'].nunique()
```

```python
no_of_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
no_of_ratings.rename(columns={'rating':'num_ratings'},inplace=True)
no_of_ratings
```

```python
final_ratings = ratings_with_books.merge(no_of_ratings,on='title')
final_ratings
```

### `num_ratings` is being set to >=50

```python
final_ratings = final_ratings[final_ratings['num_ratings']>=50]
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_ratings
```

```python
#final_ratings['title'] = final_ratings['title'].str.replace("\\","")
#final_ratings['title'] = final_ratings['title'].str.replace('"','')
```

```python
#book_pivot = final_ratings.pivot(columns='title',index='user_id',values='rating')
```

```python
book_pivot = final_ratings.pivot(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```
